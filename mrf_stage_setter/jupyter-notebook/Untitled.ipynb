```python
import os
import struct

def gzip_isize(path: str) -> int:
    with open(path, "rb") as f:
        f.seek(-4, os.SEEK_END)
        (isize,) = struct.unpack("<I", f.read(4))  # little-endian uint32
    return isize

print(gzip_isize("D:\\payer_mrf_2\\data\\Blue-Cross-and-Blue-Shield-of-Illinois\\2026-02-29\\_ingested_2025-12-14_Blue-Cross-and-Blue-Shield-of-Illinois_HMO-Illinois_in-network-rates.json.gz"))
```
